```python
import pandas as pd
from google.colab import drive
from dateutil.relativedelta import relativedelta
drive.mount("/content/drive")
```

```python
stores = pd.read_csv("/content/drive/MyDrive/Side Project/global sales/Raw Datasets/Stores.csv")
sales = pd.read_csv("/content/drive/MyDrive/Side Project/global sales/Raw Datasets/Sales.csv")
products = pd.read_csv("/content/drive/MyDrive/Side Project/global sales/Raw Datasets/Products.csv")
customers = pd.read_csv("/content/drive/MyDrive/Side Project/global sales/Raw Datasets/Customers.csv",encoding_errors='backslashreplace')
```

```python
customers["Birthday"] = pd.to_datetime(customers["Birthday"], format="%m/%d/%Y")
sales["Order Date"] =pd.to_datetime(sales["Order Date"], format="%m/%d/%Y")
customers["Age"] = customers.apply(lambda x: relativedelta(pd.to_datetime("1/3/2021"), x.Birthday).years, axis=1)
products["Unit Cost USD"] = products["Unit Cost USD"].str.replace(",","")
products["Unit Cost USD"] = products["Unit Cost USD"].str.replace("$","").astype(float)
products["Unit Price USD"] = products["Unit Price USD"].str.replace(",","")
products["Unit Price USD"] = products["Unit Price USD"].str.replace("$","").astype(float)
```

```python
display(stores.isnull().sum())
display(sales.isnull().sum())
display(products.isnull().sum())
display(customers.isnull().sum())
```

```python
stores["Square Meters"] = stores["Square Meters"].fillna(0) # Online store
customers["State Code"] = customers["State Code"].fillna("NAP") # a new state short-form
sales = sales.drop(columns=["Delivery Date"])
```

```python
sales = pd.merge(sales, products[['ProductKey','Unit Cost USD','Unit Price USD']], how="inner", left_on="ProductKey", right_on="ProductKey")
sales["Total Cost USD"] = sales["Quantity"] * sales["Unit Cost USD"]
sales["Total Price USD"] = sales["Quantity"] * sales["Unit Price USD"]
sales = sales.drop(["Unit Cost USD", "Unit Price USD"],axis=1)
```

```python
display(stores.duplicated().sum())
display(sales.duplicated().sum())
display(products.duplicated().sum())
display(customers.duplicated().sum())
```
