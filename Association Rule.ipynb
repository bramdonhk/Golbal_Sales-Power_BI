```python
import pandas as pd
from google.colab import drive

from dateutil.relativedelta import relativedelta
from mlxtend.preprocessing import TransactionEncoder

drive.mount("/content/drive")
```

```python
sales = pd.read_csv("/content/drive/MyDrive/Side Project/global sales/PowerBI/sales.csv") # cleaned datasets
products = pd.read_csv("/content/drive/MyDrive/Side Project/global sales/PowerBI/products.csv")
```

```python
record = pd.merge(sales, products, how="inner", left_on="ProductKey", right_on="ProductKey")
```

```python
category = record[["Order Number", "Category"]]
category = category.drop_duplicates()

product = record[["Order Number", "Product Name"]]
product = product.drop_duplicates()
```

```python
category["Category"] = category.groupby(['Order Number'])['Category'].transform(lambda x : ','.join(x))
category = category.drop_duplicates()
category["Category"] = category["Category"].astype("str")
category = category["Category"].str.split(",", expand=True)
```

```python
category
```

```python
product
```

```python
category_list = category.stack().groupby(level=0).apply(list).tolist()
te = TransactionEncoder()
category_tf = te.fit_transform(category_list)
category = pd.DataFrame(category_tf, columns=te.columns_)
```

```python
from mlxtend.frequent_patterns import apriori

frequent_itemsets = apriori(category,min_support=0.01, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets.sort_values(by="support",ascending=False,inplace=True)
frequent_itemsets[frequent_itemsets['length'] >= 3]
```

```python
from mlxtend.frequent_patterns import association_rules
association_rules = association_rules(frequent_itemsets,metric="confidence", min_threshold=0.5)
association_rules.sort_values(by="leverage",ascending=False,inplace=True)
association_rules
```

# **Subcategory**

```python
subcategory = record[["Order Number", "Subcategory"]]
subcategory = subcategory.drop_duplicates()

subcategory["Subcategory"] = subcategory.groupby(['Order Number'])["Subcategory"].transform(lambda x : ','.join(x))
subcategory = subcategory.drop_duplicates()
subcategory["Subcategory"] = subcategory["Subcategory"].astype("str")
subcategory = subcategory["Subcategory"].str.split(",", expand=True)
```

```python
subcategory_list = subcategory.stack().groupby(level=0).apply(list).tolist()
te = TransactionEncoder()
subcategory_tf = te.fit_transform(subcategory_list)
subcategory = pd.DataFrame(subcategory_tf, columns=te.columns_)
```

```python
from mlxtend.frequent_patterns import apriori

frequent_itemsets_subcategory = apriori(subcategory,min_support=0.001, use_colnames=True)
frequent_itemsets_subcategory['length'] = frequent_itemsets_subcategory['itemsets'].apply(lambda x: len(x))


```

```python
display(frequent_itemsets_subcategory[frequent_itemsets_subcategory['length'] >= 2])
frequent_itemsets_subcategory.sort_values(by="support",ascending=False,inplace=True)
```

```python
from mlxtend.frequent_patterns import association_rules
association_rules_subccategory = association_rules(frequent_itemsets_subcategory,metric="confidence", min_threshold=0.1)
association_rules_subccategory.sort_values(by="leverage",ascending=False,inplace=True)
```
